```python
from sklearn.cluster import KMeans
data = list(zip(df['writing score'], df['math score']))

import warnings

kmeans = KMeans(n_clusters=3).fit(data)
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method between Writing and Math Scores')

plt.show()
```

```python
from sklearn.cluster import KMeans
df.columns = df.columns.to_series().apply(lambda x: x.strip())
x = df.iloc[:,5:] # 1t for rows and second for columns
x
kmeans = KMeans(3)
kmeans.fit(x)
identified_clusters = kmeans.fit_predict(x)
identified_clusters
data_with_clusters = df.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['math score'],data_with_clusters['reading score'],c=data_with_clusters['Clusters'],cmap='rainbow')
```
